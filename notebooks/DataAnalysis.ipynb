## 6.**Define** Function for Displaying Basic Statistics

```
def display_basic_statistics(df, name):
    print(f"Basic Statistics for {name}:")
    df.describe().show()
```

## 7: Defining Function for Displaying **Correlations**

```
from pyspark.sql.functions import corr

def display_correlations(df, columns, name):
    print(f"Correlations in {name}:")
    for col1 in columns:
        for col2 in columns:
            if col1 != col2:
                print(f"Correlation between {col1} and {col2}: ", end="")
                df.select(corr(col1, col2)).show()
```

## 10: Define Function for Comprehensive Data Visualization

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_data(df, name):
    pd_df = df.toPandas()
    
    # Ensure 'Close' is numeric for plotting
    pd_df['Close'] = pd.to_numeric(pd_df['Close'], errors='coerce')
    
    # Time Series Plot
    plt.figure(figsize=(10, 6))
    plt.plot(pd_df['Date'], pd_df['Close'], label='Close Price')
    plt.title(f'{name} Closing Prices Over Time')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()

    # Histogram
    sns.histplot(pd_df['Close'], bins=30, kde=True)
    plt.title(f'Distribution of {name} Closing Prices')
    plt.show()

    # Box Plot
    sns.boxplot(x=pd_df['Close'])
    plt.title(f'Box Plot of {name} Closing Prices')
    plt.show()
```

## 11: Comprehensive Data Analysis and Visualization Execution

```
# Assuming 'dataframes' is a dictionary with dataset names as keys and DataFrames as values
for name, df in dataframes.items():
    # Display Basic Statistics
    display_basic_statistics(df, name)
    
    # Assuming you want to check correlations between 'Open', 'Close', 'High', 'Low'
    display_correlations(df, ['Open', 'Close', 'High', 'Low'], name)
    
    # Visualize Data
    visualize_data(df, name)
```
